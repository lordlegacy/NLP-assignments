```python
from selenium import webdriver
from bs4 import BeautifulSoup

# URL of the webpage you want to download
url = "https://news.lockheedmartin.com/goes-u-ships-to-florida"

# Set up the WebDriver (make sure you have the appropriate driver installed, e.g., chromedriver for Chrome)
driver = webdriver.Chrome()
driver.get(url)

# Wait for the page to load (you might need to adjust the time based on the page you are working with)
driver.implicitly_wait(20)

# Get the page source after it's fully loaded
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Extract and print the text content of the page
text_content = soup.get_text(separator='\n').strip()
print(text_content)
```

```python


body_element = soup.find(class_="wd_body wd_news_body")

# Extract and print the text content of the body element
if body_element:
    body_text = body_element.get_text(separator='\n').strip()
    print(body_text)
else:
    print("No element with class 'wd_body wd_news_body' found.")
```

```python
# Perform POS tagging
pos_tags = nltk.pos_tag(words)

# Extract person names based on POS tags
persons = [word for word, pos in pos_tags if pos == 'NNP']  # NNP is the POS tag for proper nouns (which often include names)

# Print person names
for person in persons:
    print(person)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet


# Assuming 'filtered_words' contains the non-stop words
lemmatizer = WordNetLemmatizer()

# Lemmatize the words
lemmatized_words = [lemmatizer.lemmatize(word, pos=wordnet.VERB) for word in filtered_words]

# Print each lemmatized word
for lemmatized_word in lemmatized_words:
    print(lemmatized_word)
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt





# Remove punctuation and convert to lowercase
words = [word.lower() for word in words if word.isalpha()]

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Lemmatize the words
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# Create a frequency distribution before lemmatization
fdist_before = FreqDist(filtered_words)

# Create a frequency distribution after lemmatization
fdist_after = FreqDist(lemmatized_words)

# Plot the frequency distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
fdist_before.plot(30, cumulative=False)
plt.title('Frequency Distribution Before Lemmatization')

plt.subplot(1, 2, 2)
fdist_after.plot(30, cumulative=False)
plt.title('Frequency Distribution After Lemmatization')

plt.show()
```
